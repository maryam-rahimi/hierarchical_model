```python
# =========================================================
#  HARMONI-HELM-ME: Full TUAB Spatial-Hierarchical EEG Graph Transformer
# =========================================================
# - Full dataset (normal + all abnormal subtypes)
# - Connectivity-based dynamic graphs (wPLI/corr)
# - Hierarchical encoder (electrode â†’ region â†’ global)
# - Multi-expert transformer (subtype-specific experts)
# - Transparent training logs (gate entropy)
# =========================================================

import os, glob, random, warnings, time, numpy as np, torch
import torch.nn as nn, torch.nn.functional as F
from tqdm import tqdm
from torch_geometric.data import Data
from torch_geometric.utils import dense_to_sparse
from torch_geometric.nn import GCNConv, SAGPooling, global_mean_pool
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
import mne
from mne_connectivity import spectral_connectivity_epochs

warnings.filterwarnings("ignore")
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
torch.manual_seed(42)
np.random.seed(42)
random.seed(42)

# =====================================================
#  DATA LOADING (Full)
# =====================================================
root_dir = r"C:\Users\mrahi\tuh\edf_downloads"

subtype_map = {
    "epileptiform": 0, "focal_slowing": 1, "diffuse_slowing": 2,
    "asymmetry": 3, "burst_suppression": 4, "beta_excess": 5, "unspecified": 6
}

def load_and_preprocess_edf(file_path):
    """Load and preprocess EDF file."""
    try:
        raw = mne.io.read_raw_edf(file_path, preload=True, verbose=False)
        raw.pick_types(eeg=True)
        raw.filter(1., 40., verbose=False)
        raw.resample(250, verbose=False)
        return raw
    except Exception as e:
        print(f"âš ï¸ Could not load {file_path}: {e}")
        return None

def calculate_connectivity_matrix(raw, max_segments=6):
    """Compute wPLI or fallback correlation."""
    data = raw.get_data()
    sfreq = raw.info['sfreq']
    n_channels, n_samples = data.shape
    seg_len = n_samples // max_segments
    matrices = []
    for i in range(max_segments):
        start, end = i * seg_len, (i + 1) * seg_len
        if end > n_samples: break
        segment = data[:, start:end]
        if segment.shape[1] < 50: continue
        try:
            conn = spectral_connectivity_epochs(
                [segment[np.newaxis, :, :]],
                method="wpli", sfreq=sfreq, fmin=4, fmax=30,
                faverage=True, verbose=False
            ).get_data(output="dense")
            A = conn[0, :, :, 0]
        except Exception:
            A = np.corrcoef(segment)
        np.fill_diagonal(A, 0)
        A = np.nan_to_num(A)
        X = np.vstack([
            segment.mean(axis=1),
            segment.std(axis=1),
            np.ptp(segment, axis=1)
        ]).T
        X = (X - X.mean(0)) / (X.std(0) + 1e-8)
        matrices.append((A, X))
    return matrices

# -------------------------------
# Full TUAB Discovery
# -------------------------------
file_paths, labels_lvl1, labels_lvl2 = [], [], []

normal_dir = os.path.join(root_dir, "normal")
normal_files = glob.glob(os.path.join(normal_dir, "*.edf"))
for fp in normal_files:
    file_paths.append(fp)
    labels_lvl1.append(0)
    labels_lvl2.append(-1)

abnormal_dir = os.path.join(root_dir, "abnormal")
for subtype in os.listdir(abnormal_dir):
    subtype_dir = os.path.join(abnormal_dir, subtype)
    if not os.path.isdir(subtype_dir): continue
    edf_files = glob.glob(os.path.join(subtype_dir, "*.edf"))
    for fp in edf_files:
        file_paths.append(fp)
        labels_lvl1.append(1)
        labels_lvl2.append(subtype_map.get(subtype, 6))

print(f"ðŸ“Š TUAB discovery complete | Total={len(file_paths)} | Normal={labels_lvl1.count(0)} | Abnormal={labels_lvl1.count(1)}")

# =====================================================
# 2ï¸âƒ£ BUILD GRAPH DATASET
# =====================================================
def build_dynamic_graphs(file_paths, labels1, labels2, max_segments=6):
    graphs = []
    for fp, y1, y2 in tqdm(zip(file_paths, labels1, labels2),
                           total=len(file_paths), desc="Building EEG Graphs"):
        raw = load_and_preprocess_edf(fp)
        if raw is None: continue
        for A, X in calculate_connectivity_matrix(raw, max_segments):
            edge_index, edge_weight = dense_to_sparse(torch.tensor(A, dtype=torch.float))
            g = Data(
                x=torch.tensor(X, dtype=torch.float),
                edge_index=edge_index, edge_weight=edge_weight,
                y=torch.tensor(y1).long(), y2=torch.tensor(y2).long()
            )
            graphs.append(g)
    return graphs

graphs = build_dynamic_graphs(file_paths, labels_lvl1, labels_lvl2)
train_graphs, test_graphs = train_test_split(
    graphs, test_size=0.2,
    stratify=[g.y.item() for g in graphs],
    random_state=42
)
print(f"âœ… Graph dataset ready | Train={len(train_graphs)} | Test={len(test_graphs)}")

# =====================================================
# 3ï¸âƒ£ HIERARCHICAL GRAPH ENCODER
# =====================================================
class HierarchicalGraphEncoder(nn.Module):
    def __init__(self, in_channels, hidden_channels, ratio=0.5):
        super().__init__()
        self.gcn1 = GCNConv(in_channels, hidden_channels)
        self.pool1 = SAGPooling(hidden_channels, ratio)
        self.gcn2 = GCNConv(hidden_channels, hidden_channels)
        self.pool2 = SAGPooling(hidden_channels, ratio)
        self.gcn3 = GCNConv(hidden_channels, hidden_channels)
        self.trace = {}

    def forward(self, x, edge_index, batch=None):
        if batch is None:
            batch = torch.zeros(x.size(0), dtype=torch.long, device=x.device)
        x1 = F.relu(self.gcn1(x, edge_index))
        x1p = global_mean_pool(x1, batch)
        self.trace["electrode"] = x1.detach().mean(0).cpu()
        x2, ei2, _, b2, _, _ = self.pool1(x1, edge_index, None, batch)
        x2 = F.relu(self.gcn2(x2, ei2))
        x2p = global_mean_pool(x2, b2)
        self.trace["region"] = x2.detach().mean(0).cpu()
        x3, ei3, _, b3, _, _ = self.pool2(x2, ei2, None, b2)
        x3 = F.relu(self.gcn3(x3, ei3))
        x3p = global_mean_pool(x3, b3)
        self.trace["global"] = x3.detach().mean(0).cpu()
        return x1p, x2p, x3p

# =====================================================
#  HELM-ME MODEL
# =====================================================
class HARMONI_HELM_ME(nn.Module):
    def __init__(self, in_channels=3, hidden_channels=128,
                 num_heads=4, num_layers=2, dropout=0.3, n_subtypes=7):
        super().__init__()
        self.encoder = HierarchicalGraphEncoder(in_channels, hidden_channels)
        self.norm = nn.LayerNorm(hidden_channels)
        enc_layer = nn.TransformerEncoderLayer(
            d_model=hidden_channels, nhead=num_heads,
            dim_feedforward=hidden_channels*2,
            dropout=dropout, activation="gelu", batch_first=True)
        self.transformer = nn.TransformerEncoder(enc_layer, num_layers=num_layers)
        self.dropout = nn.Dropout(dropout)
        self.experts = nn.ModuleList([nn.Linear(hidden_channels, 2) for _ in range(n_subtypes)])
        self.fc_gate = nn.Linear(hidden_channels, n_subtypes)
        self.fc_fuse = nn.Linear(n_subtypes*2, 2)
        self.fc_subtype = nn.Linear(hidden_channels, n_subtypes)
        self.last_gates = None
        self.last_confidences = None

    def forward(self, graph_seq, explain=False):
        h1s, h2s, h3s = [], [], []
        for g in graph_seq:
            h1, h2, h3 = self.encoder(g.x.to(device), g.edge_index.to(device))
            h1s.append(h1); h2s.append(h2); h3s.append(h3)
        h3_seq = self.norm(torch.stack(h3s, dim=1))
        h_out = self.transformer(h3_seq)
        h_last = self.dropout(h_out.mean(1))
        gates = F.softmax(self.fc_gate(h_last), dim=-1)
        expert_logits = [exp(h_last) for exp in self.experts]
        expert_stack = torch.cat(expert_logits, dim=-1)
        fused_logits = self.fc_fuse(expert_stack)
        subtype_logits = self.fc_subtype(h_last)

        if explain:
            self.last_gates = gates.detach().cpu()
            self.last_confidences = {
                "subtype_probs": F.softmax(subtype_logits, dim=-1).detach().cpu().numpy(),
                "global_probs": F.softmax(fused_logits, dim=-1).detach().cpu().numpy()
            }

        return {"fused": fused_logits, "experts": expert_logits,
                "subtype": subtype_logits, "gates": gates}, [torch.stack(h1s), torch.stack(h2s), torch.stack(h3s)]

# =====================================================
#  LOSS
# =====================================================
def helm_me_loss(preds, y1, y2, h_levels,
                 Î»_global=1.0, Î»_expert=0.4, Î»_sem=0.3, Î»_consistency=0.2):
    L_global = F.cross_entropy(preds["fused"], y1)
    L_expert = 0
    for i, logit in enumerate(preds["experts"]):
        mask = (y2 == i) | (y1 == 0)
        if mask.any():
            L_expert += F.cross_entropy(logit[mask], y1[mask])
    L_expert /= len(preds["experts"])
    L_sem = F.cross_entropy(preds["subtype"], y2.clamp(min=0))
    reg = sum(1 - F.cosine_similarity(h_levels[i].mean(1),
                                      h_levels[i+1].mean(1)).mean()
              for i in range(len(h_levels)-1))
    return Î»_global*L_global + Î»_expert*L_expert + Î»_sem*L_sem + Î»_consistency*reg

# =====================================================
# TRAINING & EVAL
# =====================================================
def train_dynamic(model, graphs, epochs=15, lr=5e-4):
    opt = torch.optim.AdamW(model.parameters(), lr=lr, weight_decay=1e-4)
    model.train()
    start = time.time()
    for ep in range(epochs):
        random.shuffle(graphs)
        total = 0
        for g in tqdm(graphs, desc=f"Epoch {ep+1}/{epochs}"):
            opt.zero_grad()
            preds, h = model([g])
            y1, y2 = g.y.to(device).unsqueeze(0), g.y2.to(device).unsqueeze(0)
            loss = helm_me_loss(preds, y1, y2, h)
            loss.backward(); opt.step(); total += loss.item()
        print(f"Epoch {ep+1}: mean loss = {total/len(graphs):.4f}")

        if (ep + 1) % 3 == 0 or ep == epochs - 1:
            model.eval()
            g = random.choice(graphs)
            model([g], explain=True)
            gates = model.last_gates.mean(0).numpy()
            entropy = -np.sum(gates * np.log(gates + 1e-8))
            print(f"ðŸ” Epoch {ep+1} Transparency | Gate Entropy={entropy:.3f}")
            if torch.cuda.is_available():
                print(f"ðŸ’¾ GPU Memory Used: {torch.cuda.memory_allocated()/1e9:.2f} GB")
            model.train()

    print(f"â± Training complete in {(time.time()-start)/60:.1f} min")

def evaluate(model, graphs):
    model.eval(); preds, labels = [], []
    with torch.no_grad():
        for g in graphs:
            out, _ = model([g])
            preds.append(out["fused"].argmax(-1).item())
            labels.append(g.y.item())
    acc = accuracy_score(labels, preds)
    print(f"\nâœ… Test Accuracy: {acc:.4f}")
    print(classification_report(labels, preds, target_names=["Normal","Abnormal"]))
    cm = confusion_matrix(labels, preds)
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
                xticklabels=["Normal","Abnormal"],
                yticklabels=["Normal","Abnormal"])
    plt.title("HELM-ME Confusion Matrix"); plt.show()

# =====================================================
# 7ï¸âƒ£ RUN
# =====================================================
if __name__ == "__main__":
    model = HARMONI_HELM_ME().to(device)
    print(f"ðŸš€ Starting full TUAB training on {len(train_graphs)} graphs...")
    train_dynamic(model, train_graphs, epochs=15, lr=5e-4)
    evaluate(model, test_graphs)
    torch.save(model.state_dict(), "harmoni_helm_me_full.pt")
    print("ðŸ’¾ Model saved â†’ harmoni_helm_me_full.pt")
```

```python
# =====================================================
# 8ï¸âƒ£ HYPERPARAMETER TUNING (SAFE + CORRECTED)
# =====================================================
if __name__ == "__main__":
    import pandas as pd
    from sklearn.metrics import accuracy_score, balanced_accuracy_score, f1_score

    print("\nðŸš€ Starting HARMONI-HELM-ME hyperparameter tuning ...")

    def quick_eval(model, graphs):
        """Quick evaluation on small test subset."""
        model.eval()
        y_true, y_pred = [], []
        with torch.no_grad():
            subset = random.sample(graphs, min(200, len(graphs)))
            for g in subset:
                out, _ = model([g])
                y_true.append(g.y.item())
                y_pred.append(out["fused"].argmax(-1).item())
        return (
            accuracy_score(y_true, y_pred),
            balanced_accuracy_score(y_true, y_pred),
            f1_score(y_true, y_pred)
        )

    def train_and_eval(cfg):
        """Train a smaller version of HARMONI_HELM_ME for tuning."""
        model = HARMONI_HELM_ME(
            hidden_channels=cfg["hidden_channels"],
            num_heads=cfg["num_heads"],
            num_layers=cfg["num_layers"],
            dropout=cfg["dropout"]
        ).to(device)

        opt = torch.optim.AdamW(model.parameters(), lr=cfg["lr"], weight_decay=cfg["weight_decay"])
        model.train()
        for ep in range(cfg["epochs"]):
            random.shuffle(train_graphs)
            for g in random.sample(train_graphs, 300):  # subsample for speed
                opt.zero_grad()
                preds, h = model([g])
                y1, y2 = g.y.to(device).unsqueeze(0), g.y2.to(device).unsqueeze(0)
                loss = helm_me_loss(
                    preds, y1, y2, h,
                    Î»_global=cfg["Î»_global"],
                    Î»_expert=cfg["Î»_expert"],
                    Î»_sem=0.3, Î»_consistency=0.2
                )
                loss.backward()
                torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
                opt.step()

        acc, bac, f1 = quick_eval(model, test_graphs)
        return acc, bac, f1

    # --------------------------------------------------
    # VALID PARAMETER SPACE
    # --------------------------------------------------
    valid_pairs = [(128, 4), (192, 6), (192, 8), (256, 8)]
    space = {
        "num_layers": [2, 3],
        "dropout": [0.2, 0.3, 0.4],
        "lr": [1e-3, 5e-4, 3e-4],
        "weight_decay": [1e-4, 5e-5],
        "Î»_global": [1.0, 1.1, 1.2],
        "Î»_expert": [0.3, 0.4, 0.5],
        "epochs": [4]
    }

    # --------------------------------------------------
    # RUN TUNING
    # --------------------------------------------------
    results = []
    for i in range(10):  # try 10 random configs
        cfg = {k: random.choice(v) for k, v in space.items()}
        cfg["hidden_channels"], cfg["num_heads"] = random.choice(valid_pairs)
        print(f"\nðŸ§ª Run {i+1}/10: {cfg}")

        acc, bac, f1 = train_and_eval(cfg)
        cfg.update({"acc": acc, "bac": bac, "f1": f1})
        results.append(cfg)

        pd.DataFrame(results).to_csv("harmoni_helm_tuning_results.csv", index=False)
        print(f"â†’ ACC={acc:.3f}, BAC={bac:.3f}, F1={f1:.3f}")

    # --------------------------------------------------
    # SUMMARY
    # --------------------------------------------------
    df = pd.DataFrame(results).sort_values("bac", ascending=False)
    print("\nðŸ† Top Configurations by Balanced Accuracy:")
    print(df.head(5))
    df.to_csv("harmoni_helm_tuning_results.csv", index=False)
    print("ðŸ“ Saved â†’ harmoni_helm_tuning_results.csv")
```

```python
# =====================================================
# 9ï¸âƒ£ OPTIMIZER + ACTIVATION TUNING
# =====================================================
if __name__ == "__main__":
    import torch.optim as optim
    import pandas as pd

    print("\nðŸš€ Starting optimization & activation tuning...")

    # --------------------------
    # Define candidate options
    # --------------------------
    optimizers = ["AdamW", "Adam", "RAdam", "SGD", "Lion"]
    activations = ["gelu", "relu", "silu", "mish"]
    schedulers = ["none", "cosine", "plateau"]

    # Quick small eval subset
    small_train = random.sample(train_graphs, min(300, len(train_graphs)))
    small_test = random.sample(test_graphs, min(200, len(test_graphs)))

    results = []

    for opt_name in optimizers:
        for act_name in activations:
            for sch_name in schedulers:
                cfg = {
                    "optimizer": opt_name,
                    "activation": act_name,
                    "scheduler": sch_name
                }
                print(f"\nðŸ§  Testing {cfg}")

                # ----------- Model -------------
                act_fn = dict(gelu=F.gelu, relu=F.relu, silu=F.silu,
                              mish=lambda x: x * torch.tanh(F.softplus(x)))[act_name]
                model = HARMONI_HELM_ME(
                    hidden_channels=192, num_heads=8,
                    num_layers=2, dropout=0.3
                ).to(device)

                # Override transformer activation dynamically
                for m in model.modules():
                    if isinstance(m, nn.TransformerEncoderLayer):
                        m.activation = act_fn

                # ----------- Optimizer ----------
                if opt_name == "AdamW":
                    opt = optim.AdamW(model.parameters(), lr=5e-4, weight_decay=1e-4)
                elif opt_name == "Adam":
                    opt = optim.Adam(model.parameters(), lr=5e-4)
                elif opt_name == "RAdam":
                    opt = optim.RAdam(model.parameters(), lr=5e-4)
                elif opt_name == "SGD":
                    opt = optim.SGD(model.parameters(), lr=1e-3, momentum=0.9)
                elif opt_name == "Lion":
                    try:
                        from torch.optim import Lion
                        opt = Lion(model.parameters(), lr=5e-4)
                    except Exception:
                        print("âš ï¸ Lion optimizer not available; skipping.")
                        continue

                # ----------- Scheduler ----------
                if sch_name == "cosine":
                    scheduler = optim.lr_scheduler.CosineAnnealingLR(opt, T_max=5)
                elif sch_name == "plateau":
                    scheduler = optim.lr_scheduler.ReduceLROnPlateau(opt, patience=2)
                else:
                    scheduler = None

                # ----------- Training -----------
                model.train()
                for ep in range(4):
                    total_loss = 0
                    for g in small_train:
                        opt.zero_grad()
                        preds, h = model([g])
                        y1, y2 = g.y.to(device).unsqueeze(0), g.y2.to(device).unsqueeze(0)
                        loss = helm_me_loss(preds, y1, y2, h)
                        loss.backward()
                        torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
                        opt.step()
                        total_loss += loss.item()
                    if scheduler:
                        if sch_name == "plateau":
                            scheduler.step(total_loss)
                        else:
                            scheduler.step()
                    print(f"  Epoch {ep+1}/4 | Loss={total_loss/len(small_train):.4f}")

                # ----------- Evaluation ----------
                acc, bac, f1 = 0, 0, 0
                y_true, y_pred = [], []
                model.eval()
                with torch.no_grad():
                    for g in small_test:
                        out, _ = model([g])
                        y_true.append(g.y.item())
                        y_pred.append(out["fused"].argmax(-1).item())
                acc = accuracy_score(y_true, y_pred)
                bac = balanced_accuracy_score(y_true, y_pred)
                f1 = f1_score(y_true, y_pred)
                print(f"â†’ ACC={acc:.3f}, BAC={bac:.3f}, F1={f1:.3f}")

                cfg.update({"acc": acc, "bac": bac, "f1": f1})
                results.append(cfg)
                pd.DataFrame(results).to_csv("harmoni_helm_optimizer_tuning.csv", index=False)

    # --------------------------
    # Summary
    # --------------------------
    df = pd.DataFrame(results).sort_values("bac", ascending=False)
    print("\nðŸ† Top Optimizer/Activation Settings:")
    print(df.head(5))
    df.to_csv("harmoni_helm_optimizer_tuning.csv", index=False)
    print("ðŸ“ Saved â†’ harmoni_helm_optimizer_tuning.csv")
```

```python
# =========================================================
#  HARMONI-HELM-ME: FULL ABLATION STUDY (SELF-CONTAINED)
# =========================================================
# Models included:
#   - Baseline GCN
#   - Baseline Transformer
#   - Full HARMONI-HELM-ME
# =========================================================

import random, time, numpy as np, torch
import torch.nn as nn, torch.nn.functional as F
from tqdm import tqdm
from torch_geometric.nn import (
    GCNConv, SAGPooling, global_mean_pool
)
from sklearn.metrics import (
    accuracy_score, balanced_accuracy_score,
    classification_report, confusion_matrix, f1_score
)
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# =========================================================
#  1. BASELINE 1 â€” Simple GCN
# =========================================================
class Baseline_GCN(nn.Module):
    def __init__(self, in_channels=3, hidden_channels=128, num_classes=2):
        super().__init__()
        self.conv1 = GCNConv(in_channels, hidden_channels)
        self.conv2 = GCNConv(hidden_channels, hidden_channels)
        self.lin = nn.Linear(hidden_channels, num_classes)

    def forward(self, graphs):
        g = graphs[0]
        x, edge_index = g.x.to(device), g.edge_index.to(device)
        batch = torch.zeros(x.size(0), dtype=torch.long, device=device)
        x = F.relu(self.conv1(x, edge_index))
        x = F.dropout(x, p=0.3, training=self.training)
        x = F.relu(self.conv2(x, edge_index))
        x = global_mean_pool(x, batch)
        return self.lin(x)

# =========================================================
#  2. BASELINE 2 â€” Graph Transformer (no hierarchy)
# =========================================================
class Baseline_Transformer(nn.Module):
    def __init__(self, in_channels=3, hidden_channels=128,
                 num_heads=4, num_layers=2, dropout=0.3):
        super().__init__()
        self.gcn = GCNConv(in_channels, hidden_channels)
        self.norm = nn.LayerNorm(hidden_channels)
        enc_layer = nn.TransformerEncoderLayer(
            d_model=hidden_channels, nhead=num_heads,
            dim_feedforward=hidden_channels*2,
            dropout=dropout, batch_first=True
        )
        self.transformer = nn.TransformerEncoder(enc_layer, num_layers=num_layers)
        self.fc = nn.Linear(hidden_channels, 2)

    def forward(self, graphs):
        h_seq = []
        for g in graphs:
            x, edge_index = g.x.to(device), g.edge_index.to(device)
            batch = torch.zeros(x.size(0), dtype=torch.long, device=device)
            h = F.relu(self.gcn(x, edge_index))
            h = global_mean_pool(h, batch)
            h_seq.append(h)
        h_seq = torch.stack(h_seq, dim=1)
        h_seq = self.norm(h_seq)
        out = self.transformer(h_seq).mean(1)
        return self.fc(out)

# =========================================================
#  3. FULL MODEL â€” HARMONI-HELM-ME
# =========================================================
class HierarchicalGraphEncoder(nn.Module):
    def __init__(self, in_channels, hidden_channels, ratio=0.5):
        super().__init__()
        self.gcn1 = GCNConv(in_channels, hidden_channels)
        self.pool1 = SAGPooling(hidden_channels, ratio)
        self.gcn2 = GCNConv(hidden_channels, hidden_channels)
        self.pool2 = SAGPooling(hidden_channels, ratio)
        self.gcn3 = GCNConv(hidden_channels, hidden_channels)

    def forward(self, x, edge_index, batch=None):
        if batch is None:
            batch = torch.zeros(x.size(0), dtype=torch.long, device=x.device)
        x1 = F.relu(self.gcn1(x, edge_index))
        x2, ei2, _, b2, _, _ = self.pool1(x1, edge_index, None, batch)
        x2 = F.relu(self.gcn2(x2, ei2))
        x3, ei3, _, b3, _, _ = self.pool2(x2, ei2, None, b2)
        x3 = F.relu(self.gcn3(x3, ei3))
        h1 = global_mean_pool(x1, batch)
        h2 = global_mean_pool(x2, b2)
        h3 = global_mean_pool(x3, b3)
        return h1, h2, h3

class HARMONI_HELM_ME(nn.Module):
    def __init__(self, in_channels=3, hidden_channels=128,
                 num_heads=4, num_layers=2, dropout=0.3, n_subtypes=7):
        super().__init__()
        self.encoder = HierarchicalGraphEncoder(in_channels, hidden_channels)
        self.norm = nn.LayerNorm(hidden_channels)
        enc_layer = nn.TransformerEncoderLayer(
            d_model=hidden_channels, nhead=num_heads,
            dim_feedforward=hidden_channels*2,
            dropout=dropout, activation="gelu", batch_first=True)
        self.transformer = nn.TransformerEncoder(enc_layer, num_layers=num_layers)
        self.dropout = nn.Dropout(dropout)
        self.experts = nn.ModuleList([nn.Linear(hidden_channels, 2) for _ in range(n_subtypes)])
        self.fc_gate = nn.Linear(hidden_channels, n_subtypes)
        self.fc_fuse = nn.Linear(n_subtypes*2, 2)

    def forward(self, graph_seq):
        h1s, h2s, h3s = [], [], []
        for g in graph_seq:
            h1, h2, h3 = self.encoder(g.x.to(device), g.edge_index.to(device))
            h1s.append(h1); h2s.append(h2); h3s.append(h3)
        h3_seq = self.norm(torch.stack(h3s, dim=1))
        h_out = self.transformer(h3_seq)
        h_last = self.dropout(h_out.mean(1))
        gates = F.softmax(self.fc_gate(h_last), dim=-1)
        expert_logits = [exp(h_last) for exp in self.experts]
        expert_stack = torch.cat(expert_logits, dim=-1)
        fused_logits = self.fc_fuse(expert_stack)
        return fused_logits

# =========================================================
#  4. TRAINING AND EVALUATION
# =========================================================
def train_model(model, graphs, epochs=6, lr=1e-3):
    opt = torch.optim.AdamW(model.parameters(), lr=lr, weight_decay=1e-4)
    model.train()
    for ep in range(epochs):
        random.shuffle(graphs)
        total = 0
        for g in tqdm(graphs, desc=f"Epoch {ep+1}/{epochs}"):
            opt.zero_grad()
            logits = model([g])
            loss = F.cross_entropy(logits, g.y.to(device).unsqueeze(0))
            loss.backward()
            opt.step()
            total += loss.item()
        print(f"Epoch {ep+1}: mean loss = {total/len(graphs):.4f}")

def evaluate_model(model, graphs, name="Model"):
    model.eval()
    preds, labels = [], []
    with torch.no_grad():
        for g in graphs:
            logits = model([g])
            preds.append(logits.argmax(-1).item())
            labels.append(g.y.item())
    acc = accuracy_score(labels, preds)
    bac = balanced_accuracy_score(labels, preds)
    f1 = f1_score(labels, preds)
    print(f"\n{name} Results:")
    print(f"Accuracy: {acc:.4f}")
    print(f"Balanced Accuracy: {bac:.4f}")
    print(f"F1 Score: {f1:.4f}")
    print(classification_report(labels, preds, target_names=["Normal", "Abnormal"]))
    cm = confusion_matrix(labels, preds)
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
                xticklabels=["Normal", "Abnormal"],
                yticklabels=["Normal", "Abnormal"])
    plt.title(f"{name} Confusion Matrix")
    plt.show()
    return acc, bac, f1

# =========================================================
#  5. ABLATION RUNNER
# =========================================================
def run_ablation(train_graphs, test_graphs):
    models = {
        "Baseline_GCN": Baseline_GCN().to(device),
        "Baseline_Transformer": Baseline_Transformer().to(device),
        "HARMONI_HELM_ME": HARMONI_HELM_ME().to(device)
    }

    results = []
    for name, model in models.items():
        print("===================================")
        print(f"Training {name}")
        print("===================================")
        start = time.time()
        train_model(model, train_graphs, epochs=6, lr=1e-3)
        acc, bac, f1 = evaluate_model(model, test_graphs, name)
        elapsed = (time.time() - start) / 60
        results.append({"Model": name, "ACC": acc, "BAC": bac, "F1": f1, "Time_min": elapsed})

    df = pd.DataFrame(results)
    print("\nAblation Summary:")
    print(df)
    df.to_csv("harmoni_ablation_results.csv", index=False)
    print("Results saved to harmoni_ablation_results.csv")

# =========================================================
#  6. RUN (use already loaded graphs)
# =========================================================
# Just call this in your notebook after graphs are built:
# run_ablation(train_graphs, test_graphs)
```

```python
# =========================================================
#  ðŸ§  HARMONI Baseline Ablation (GCN + Transformer)
# =========================================================
import torch, torch.nn as nn, torch.nn.functional as F
import pandas as pd, random, time
from torch_geometric.nn import GCNConv, global_mean_pool
from sklearn.metrics import accuracy_score, balanced_accuracy_score, f1_score
from tqdm import tqdm

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# =========================================================
# Utility Functions
# =========================================================
def get_activation(name):
    name = name.lower()
    return {
        "relu": nn.ReLU(),
        "gelu": nn.GELU(),
        "silu": nn.SiLU(),
        "mish": nn.Mish()
    }.get(name, nn.ReLU())

def get_optimizer(name, params, lr=5e-4):
    name = name.lower()
    if name == "adam":  return torch.optim.Adam(params, lr=lr)
    if name == "adamw": return torch.optim.AdamW(params, lr=lr, weight_decay=1e-4)
    if name == "radam": return torch.optim.RAdam(params, lr=lr)
    return torch.optim.Adam(params, lr=lr)

# =========================================================
# Baseline Models
# =========================================================
class BaselineGCN(nn.Module):
    def __init__(self, in_ch=3, hidden=128, act="relu"):
        super().__init__()
        self.act = get_activation(act)
        self.conv1 = GCNConv(in_ch, hidden)
        self.conv2 = GCNConv(hidden, hidden)
        self.lin = nn.Linear(hidden, 2)
    def forward(self, g):
        x, edge = g.x.to(device), g.edge_index.to(device)
        x = self.act(self.conv1(x, edge))
        x = self.act(self.conv2(x, edge))
        x = global_mean_pool(x, torch.zeros(x.size(0), dtype=torch.long, device=device))
        return self.lin(x)

class BaselineTransformer(nn.Module):
    def __init__(self, in_ch=3, hidden=128, act="relu", heads=4, layers=2):
        super().__init__()
        self.embed = nn.Linear(in_ch, hidden)
        act_name = act.lower()
        if act_name not in ["relu", "gelu"]:
            act_name = "relu"
        self.encoder = nn.TransformerEncoder(
            nn.TransformerEncoderLayer(
                d_model=hidden,
                nhead=heads,
                dim_feedforward=hidden*2,
                dropout=0.2,
                activation=act_name,
                batch_first=True
            ), num_layers=layers)
        self.fc = nn.Linear(hidden, 2)
        self.custom_act = get_activation(act)
    def forward(self, g):
        x = g.x.to(device)
        h = self.custom_act(self.embed(x)).unsqueeze(0)
        h = self.encoder(h).mean(1)
        return self.fc(h)

# =========================================================
# Training & Evaluation
# =========================================================
def train_model(model, graphs, name, optimizer="RAdam", activation="relu", epochs=10, lr=5e-4):
    opt = get_optimizer(optimizer, model.parameters(), lr)
    model.to(device)
    print(f"\nðŸš€ Training {name} ({optimizer}, {activation}) for {epochs} epochs")
    start = time.time()
    for ep in range(epochs):
        model.train(); losses=[]
        random.shuffle(graphs)
        for g in graphs:
            y = g.y.to(device).unsqueeze(0)
            opt.zero_grad()
            out = model(g)
            loss = F.cross_entropy(out, y)
            if torch.isnan(loss): continue
            loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
            opt.step()
            losses.append(loss.item())
        if (ep+1)%5==0:
            print(f"  Epoch {ep+1}/{epochs} | Loss={sum(losses)/len(losses):.4f}")
    total_time = (time.time() - start) / 60
    print(f"âœ… {name} training complete in {total_time:.2f} min")
    return model, total_time

def evaluate_model(model, graphs, name):
    if len(graphs) == 0:
        print(f"âš ï¸ No test graphs found for {name}.")
        return 0, 0, 0
    model.eval(); y_true, y_pred = [], []
    with torch.no_grad():
        for g in graphs:
            y = g.y.item()
            logits = model(g)
            pred = logits.argmax(-1).item()
            y_true.append(y); y_pred.append(pred)
    acc = accuracy_score(y_true, y_pred)
    bac = balanced_accuracy_score(y_true, y_pred)
    f1  = f1_score(y_true, y_pred)
    print(f"âœ… {name} Results â†’ ACC={acc:.3f}, BAC={bac:.3f}, F1={f1:.3f}")
    return acc, bac, f1

# =========================================================
# Run Ablation
# =========================================================
def run_baseline_ablation(train_graphs, test_graphs):
    if not train_graphs or not test_graphs:
        print("âš ï¸ train_graphs or test_graphs are empty â€” please load them first.")
        return
    cfg = {"optimizer": "RAdam", "activation": "ReLU"}  # from your best run
    results = []
    models = [
        ("Vanilla_GCN", BaselineGCN(in_ch=3, hidden=128, act=cfg["activation"])),
        ("Vanilla_Transformer", BaselineTransformer(in_ch=3, hidden=128, act=cfg["activation"]))
    ]
    for name, model in models:
        model, t_min = train_model(model, train_graphs, name,
                                   optimizer=cfg["optimizer"], activation=cfg["activation"])
        acc, bac, f1 = evaluate_model(model, test_graphs, name)
        results.append({"Model": name, "ACC": acc, "BAC": bac, "F1": f1, "Time_min": t_min})
    df = pd.DataFrame(results)
    print("\nðŸ Baseline Ablation Results:\n", df)
    df.to_csv("harmoni_vanilla_ablation.csv", index=False)
    print("ðŸ“ Results saved â†’ harmoni_vanilla_ablation.csv")

# =========================================================
# Run directly in Jupyter or Python
# =========================================================
try:
    run_baseline_ablation(train_graphs, test_graphs)
except NameError:
    print("âš ï¸ Please make sure train_graphs and test_graphs are already defined.")
```
